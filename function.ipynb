```python
# function with argument and no return type
def cal(l,b): #l and b are parameter
    area = l*b
    print(area)
    
l = int(input("Enter l = ")) # global variable
b = int(input("Enter b = ")) # global variable

cal(l,b)# argument
```

```python
# function with argument and no return type
def cal(x,y):
    area = x*y
    print(area)
    
l = int(input("Enter l = ")) # global variable
b = int(input("Enter b = ")) # global variable

cal(l,b)# argument
```

```python
# function with argument and no return type
def area(x,y):
    a = x*y
    print(a)

def volume(x,y,z):
    v = x*y*z
    print(v)
    
l = int(input("Enter l = ")) # global variable
b = int(input("Enter b = ")) # global variable
h = int(input("Enter h = ")) # global variable
area(l,b)
volume(l,b,h)# argument
```
